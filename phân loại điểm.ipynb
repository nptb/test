```python
# Nhập a, b, c là hệ số của một đa thức bậc 2
a = int(input('Nhập a: '))
b = int(input('Nhập b: '))
c = int(input('Nhập c: '))

if a == 1 and b == 1:
    print('Phương trình có dạng: x^2 + x + %s = 0' %(c))

else:
    if a == -1 and b == 1:
        print('Phương trình có dạng: -x^2 + x + %s = 0' %(c))
    
    elif a == 1 and b == -1:
        print('Phương trình có dạng: x^2 - x + %s = 0' %(c))
    
    elif a == -1 and b == -1:
        print('Phương trình có dạng: -x^2 - x + %s = 0' %(c))
    
    elif a == 1 and (b != 1 or b != -1):
        print('Phương trình có dạng: x^2 + %sx = %s' %(b,c))
    
    elif a == -1 and (b != 1 or b != -1):
        print('Phương trình có dạng: -x^2 + %sx = %s' %(b,c))
    
    elif b == 1 and (a != -1 or a != 1):
        print('Phương trình có dạng: %sx^2 + x = %s' %(a,c))
    
    elif b == -1 and (a != -1 or a != 1):
        print('Phương trình có dạng: %sx^2 - x = %s' %(a,c))
    
    elif a == 0 and b == 0:
        print('c = 0')
        print('Phương trình có vô số nghiệm')
    
    elif a == 0:
        print('Phương trình có dạng: %sx + %s = 0' %(b,c))
    
    elif b == 0:
        print('Phương trình có dạng: %sx^2 + %s = 0' %(a,c))
    
    else:
        print('Phương trình có dạng: %sx^2 + %sx + %s = 0' %(a,b,c))

# Các hệ số 1, -1 thì không in hệ số

# 0 thì không in  bậc đó

# Ví dụ: a = 1, b = -3, c= 0 thì in ra là x^2 – 3x 

# In ra phương trình bậc 2 dưới dạng tự nhiên như khi viết tay đa thức ra giấy hoặc bảng ax^2 + bx +c = 0 (a,b khac 0)
```
