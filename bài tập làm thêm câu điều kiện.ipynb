```python
# Theo luật từ năm 2020 thuế suất thu nhập cá nhân được tính như sau:
ThuNhap = int(input('Nhap: '))

SoNguoiPhuThuoc = int(input('Nhap so nguoi phu thuoc: '))
MucGiamTru = 11000000
NguoiPhuThuoc = 4400000

TongGiamTru = MucGiamTru + NguoiPhuThuoc * SoNguoiPhuThuoc
Luong = ThuNhap - TongGiamTru

if Luong < 5000000 and Luong > 0:
    print('Miễn nộp thuế')

# Thuế TNCN phải đóng khi biết thu nhập tính thuế của người đó
# Thu nhập tính thuế/tháng = 5 triệu: thuế 5%.
else:
    if Luong == 5000000:
        Thue = Luong * 5 / 100
        print('Thuế phải nộp là: %s VNĐ' %(Thue))

# Trên 5 triệu đến 10 triệu: thuế 10%.
    elif Luong > 5000000 and Luong == 10000000:
        Thue = Luong * 10 / 100
        print('Thuế phải nộp là: %s VNĐ' %(Thue))
    
# Trên 10 triệu đến 18 triệu: thuế 15%.
    elif Luong > 10000000 and Luong == 18000000:
        Thue = Luong * 15 / 100
        print('Thuế phải nộp là: %s VNĐ' %(Thue))
    
# Trên 18 triệu đến 32 triệu: thuế 20%.
    elif Luong > 18000000 and Luong == 32000000:
        Thue = Luong * 20 / 100
        print('Thuế phải nộp là: %s VNĐ' %(Thue))
    
# Trên 32 triệu đến 52 triệu: thuế 25%.
    elif Luong > 32000000 and Luong == 52000000:
        Thue = Luong * 25 / 100
        print('Thuế phải nộp là: %s VNĐ' %(Thue))

# Trên 52 triệu đến 80 triệu: thuế 30%.
    elif Luong > 52000000 and Luong == 80000000:
        Thue = Luong * 30 / 100
        print('Thuế phải nộp là: %s VNĐ' %(Thue))

# Trên 80 triệu: thuế 35%
    elif Luong > 80000000:
        Thue = Luong * 35 / 100
        print('Thuế phải nộp là: %s VNĐ' %(Thue))

    else:
        print('Lỗi!!')

# Thuế TNCN phải đóng khi biết tổng thu nhập và số người phụ thuộc.

# Mức giảm trừ gia cảnh được áp dụng cho bản thân người đóng thuế là 11 triệu/tháng.

# Mức áp dụng đối với người phụ thuộc là 4,4 triệu/tháng.

# Thu nhập tính thuế thu nhập cá nhân là thu nhập còn lại sau khi tổng thu nhập đã trừ đi các tổng giảm trừ gia cảnh.
```
