```python
#Nhập vào 3 số tự nhiên
a = int(input('Nhập a: '))
b = int(input('Nhập b: '))
c = int(input('Nhập c: '))

#print(a*a , b*b , c*c)

#3 số đó có tạo nên 1 bộ số Pitago hay không?
if a*a + b*b == c*c or b*b + c*c == a*a or a*a + c*c == b*b:
    print('3 số tạo nên 1 bộ số Pitago')
else:
    print('3 số ko tạo nên 1 bộ số Pitago')
```

```python
# Tính cước điện thoại bàn của một hộ gia đình:

# Nhập: số phút gọi
CallMinutes = int(input('Nhập số phút: '))

# Phí thuê bao bắt buộc: 25k
SubscriptionFee = 25000

# 50 phút đầu tiên: 600/phút
# 0 - 50 phút: 600/phút
# Tính số tiền phải trả.
if CallMinutes <= 50:
    pay1 = CallMinutes * 600 + SubscriptionFee
    print('Số tiền phải trả là: %s VNĐ' %(pay1))

# 150 phút tiếp theo: 400/phút
# 51 - 200 phút: 400/phút
# Tính số tiền phải trả.
else:
    if CallMinutes > 50 and CallMinutes <= 200:
        pay2 = (CallMinutes - 50) * 400 + 55000
        print('Số tiền phải trả là: %s VNĐ' %(pay2))

# Sau phút 200: 200/phút
# 201 phút về sau: 200/phút
# Tính số tiền phải trả.
    else:
        pay3 = (CallMinutes - 200) * 200 + 115000
        print('Số tiền phải trả là: %s VNĐ' %(pay3))
```

```python
# Nhập 1 số tự nhiên từ 0-6
DayOfWeek = int(input('Nhập số bất kỳ từ 0-6: '))

# 0 là thứ 2
if DayOfWeek == 0:
    print('Số bạn chọn tương ứng với thứ 2.')

else:
    # 1 là thứ 3 
    if DayOfWeek == 1:
        print('Số bạn chọn tương ứng với thứ 3.')

    # 2 là thứ 4 
    elif DayOfWeek == 2:        
        print('Số bạn chọn tương ứng với thứ 4.')

    # 3 là thứ 5 
    elif DayOfWeek == 3:        
        print('Số bạn chọn tương ứng với thứ 5.')

    # 4 là thứ 6 
    elif DayOfWeek == 4:        
        print('Số bạn chọn tương ứng với thứ 6.')

    # 5 là thứ 7 
    elif DayOfWeek == 5:        
        print('Số bạn chọn tương ứng với thứ 7.')

    # 6 là chủ nhật.
    elif DayOfWeek == 6:        
        print('Số bạn chọn tương ứng với chủ nhật.')

# Ngoài phạm vi 0-6 là lỗi nhập dữ liệu đầu vào.
    else:
        print('Vui lòng chỉ nhập các số từ 0 đến 6.')
```
