```python
import turtle

tl = turtle.Turtle()

tl.pencolor('black')

tl.penup()
tl.left(90)
tl.forward(200)
tl.right(90)
tl.forward(50)

tl.pendown()

tl.fillcolor('black')
tl.begin_fill()

tl.right(36)
tl.forward(100)
tl.right(36)
tl.forward(100)
tl.right(36)
tl.forward(100)
tl.right(36)
tl.forward(100)
tl.right(36)
tl.forward(100)
tl.right(36)
tl.forward(100)
tl.right(36)
tl.forward(100)
tl.right(36)
tl.forward(100)
tl.right(36)
tl.forward(100)
tl.right(36)
tl.forward(100)
tl.left(36)

tl.end_fill()

turtle.done()
```
