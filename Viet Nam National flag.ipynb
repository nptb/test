```python
import turtle

tl = turtle.Turtle()

tl.pensize(0)
tl.pencolor('yellow')

tl.penup()
tl.backward(250)
tl.right(90)
tl.backward(200)
tl.left(90)
tl.pendown()

tl.fillcolor('red')
tl.begin_fill()

tl.forward(500)
tl.right(90)
tl.forward(300)
tl.right(90)
tl.forward(500)
tl.right(90)
tl.forward(300)

tl.end_fill()

tl.right(180)
tl.forward(100)
tl.left(90)

tl.fillcolor('yellow')
tl.begin_fill()

tl.penup()
tl.forward(150)
tl.pendown()

tl.forward(200)
tl.right(144)
tl.forward(200)
tl.right(144)
tl.forward(190)
tl.right(144)
tl.forward(190)
tl.right(144)
tl.forward(200)
tl.right(144)
tl.forward(100)

tl.end_fill()

turtle.done()
```
