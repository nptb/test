```python
x = 111
y = 87
z = 56
a =  3

#thể tích hình hộp chữ nhật
Vcn = x * y * z 
print("thể tích hình hộp chữ nhật = " + str(Vcn))

#thể tích hình lập phương
Vlp = a**3
print("thể tích hình lập phương = " + str(Vlp))

#số hình lập phương tối đa có thể chứa trong hình hộp chữ nhật
slp = Vcn//Vlp

print("vậy số lập phương tối đa có thể chứa trong hình hộp chữ nhật là " + str(slp))
```

```python
x = 111
y = 87
z = 56
a = 3

#số hình lập phương tối đa có thể chứa trong hình hộp chữ nhật
slp = (x // a) * (y // a) * (z // a)

print("vậy số lập phương tối đa có thể chứa trong hình hộp chữ nhật là " + str(slp))
```
